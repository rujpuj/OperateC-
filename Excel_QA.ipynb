```python
import pandas as pd
file_path = 'Demo Test Affiliation _Piyush.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')
head_df = df.head()
head_df
```

```python
# 1. Which faculty has more colleges?
faculty_counts = df['Faculty'].value_counts()

# 2. Which faculty has the highest NAAC accreditation count?
# Filter colleges with NAAC accreditation (status containing 'A')
naac_accredited = df[df['NAAC Status'].astype(str).str.contains('A', na=False)]
faculty_naac_counts = naac_accredited['Faculty'].value_counts()

# 3. Which faculty is most common in rural areas?
rural_faculty_counts = df[df['Urban/Rural'] == 'Rural']['Faculty'].value_counts()

# 4. Which district has more colleges?
district_counts = df['District'].value_counts()

# 5. Count of aided colleges
aided_colleges_count = df[df['ManageMent'].astype(str).str.contains('Aided', na=False)].shape[0]

# 6. List of colleges started in 2023
colleges_2023 = df[df['Affiliation_Year'] == 2023]['College/Institute Name with address']

# 7. Count of autonomous colleges
autonomous_count = df[df['Autonomous_College'].astype(str).str.contains('Yes', na=False)].shape[0]

# 8. List of colleges with A++ NAAC status
# Filter for NAAC status containing 'A++'
colleges_A_plus_plus = df[df['NAAC Status'].astype(str).str.contains('A\+\+', na=False)]['College/Institute Name with address']

# 9. Which faculty has the most colleges in urban areas?
urban_faculty_counts = df[df['Urban/Rural'] == 'Urban']['Faculty'].value_counts()

# 10. Year with the highest count of first-affiliation colleges
year_counts = df['Affiliation_Year'].value_counts()

# 11. Taluka with the highest number of colleges
taluka_counts = df['Taluka'].value_counts()

# 12. Percentage of recognized vs. affiliated colleges
recognized_count = df[df['NAAC Status'].astype(str).str.contains('Recognized', na=False)].shape[0]
affiliated_count = df[df['ManageMent'].astype(str).str.contains('Affiliated', na=False)].shape[0]
percentage_recognized = (recognized_count / len(df)) * 100
percentage_affiliated = (affiliated_count / len(df)) * 100

# 13. Average NAAC grade for colleges with different management types
# First, convert NAAC grades to numeric scores for averaging
# Assign scores: A++=4, A+=3, A=2, B+=1, others=0
grade_mapping = {'A++':4, 'A+':3, 'A':2, 'B+':1}
def get_grade_score(grade):
    if pd.isna(grade):  # Handle NaN values
        return 0
    for key in grade_mapping:
        if key in str(grade): # Convert to string before checking
            return grade_mapping[key]
    return 0

df['NAAC_Score'] = df['NAAC Status'].apply(get_grade_score)
avg_naac_by_management = df.groupby('ManageMent')['NAAC_Score'].mean()

# Now, let's print the results for each question.
print("1. Which faculty has more colleges?")
print(faculty_counts)
print("\n2. Which faculty has the highest NAAC accreditation count?")
print(faculty_naac_counts)
print("\n3. Which faculty is most common in rural areas?")
print(rural_faculty_counts)
print("\n4. Which district has more colleges?")
print(district_counts)
print("\n5. Count of aided colleges")
print(aided_colleges_count)
print("\n6. List of colleges started in 2023")
print(colleges_2023)
print("\n7. Count of autonomous colleges")
print(autonomous_count)
print("\n8. List of colleges with A++ NAAC status")
print(colleges_A_plus_plus)
print("\n9. Which faculty has the most colleges in urban areas?")
print(urban_faculty_counts)
print("\n10. Year with the highest count of first-affiliation colleges")
print(year_counts.head(1)) # Print only the top year
print("\n11. Taluka with the highest number of colleges")
print(taluka_counts.head(1)) # Print only the top taluka
print("\n12. Percentage of recognized vs. affiliated colleges")
print(f"Percentage of recognized colleges: {percentage_recognized:.2f}%")
print(f"Percentage of affiliated colleges: {percentage_affiliated:.2f}%")
print("\n13. Average NAAC grade for colleges with different management types")
print(avg_naac_by_management)
```

```python
unique_naac_status = df['NAAC Status'].unique()

df['NAAC Status'] = df['NAAC Status'].astype(str)
def get_grade_score_safe(grade):
    if not grade or grade == 'nan':
        return 0
    for key in grade_mapping:
        if key in grade:
            return grade_mapping[key]
    return 0

# Apply the safe function to the column.
df['NAAC_Score'] = df['NAAC Status'].apply(get_grade_score_safe)

# Now, recalculate the average NAAC grade for different management types.
avg_naac_by_management = df.groupby('ManageMent')['NAAC_Score'].mean()

# Show the updated result.
avg_naac_by_management.head()
```

```python
!pip install gradio
import gradio as gr

# Define a function to answer questions based on the dataset

def answer_question(question):
    question = question.lower()
    if 'most colleges' in question or 'faculty' in question:
        return 'The faculty with the most colleges is: ' + str(most_colleges_faculty) + ' with ' + str(most_colleges_count) + ' colleges.'
    elif 'district' in question:
        return 'The district with the most colleges is: ' + str(district_most_colleges) + ' with ' + str(district_counts.max()) + ' colleges.'
    elif 'aided' in question:
        return 'Number of aided colleges: ' + str(aided_colleges_count)
    elif 'colleges started in' in question:
        return 'Colleges started in 2023: ' + ', '.join(colleges_2023_list)
    elif 'autonomous' in question:
        return 'Number of autonomous colleges: ' + str(autonomous_count)
    elif 'a++' in question or 'a plus plus' in question:
        return 'Colleges with A++ NAAC status: ' + ', '.join(colleges_A_plus_plus_list)
    elif 'urban' in question:
        return 'The faculty with the most colleges in urban areas is: ' + str(most_urban_faculty) + ' with ' + str(most_urban_faculty_count) + ' colleges.'
    elif 'year' in question:
        return 'Year with the most first-affiliation colleges: ' + str(most_common_affiliation_year) + ' with ' + str(most_common_affiliation_year_count) + ' colleges.'
    elif 'taluka' in question:
        return 'Taluka with the most colleges: ' + str(most_taluka) + ' with ' + str(most_taluka_count) + ' colleges.'
    elif 'recognize' in question or 'recognized' in question:
        return 'Percentage of recognized colleges: ' + str(percentage_recognized) + '%'
    elif 'affiliated' in question:
        return 'Percentage of affiliated colleges: ' + str(percentage_affiliated) + '%'
    elif 'naac grade' in question or 'naac' in question:
        return 'Average NAAC grade scores by management: ' + str(avg_naac_by_management_dict)
    else:
        return 'Sorry, I cannot answer that question based on the dataset.'

# Create a Gradio interface
iface = gr.Interface(fn=answer_question,
                     inputs=gr.Textbox(label='Ask a question about the dataset'),
                     outputs='text',
                     title='College Dataset Q&A',
                     description='Type a question about the dataset, such as "Which faculty has the most colleges?"')

# Launch the interface
iface.launch()
```
